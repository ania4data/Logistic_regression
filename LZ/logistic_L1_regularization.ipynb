```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


def sigmoid(z):
    return 1/(1+np.exp(-z))

N=50
D=50

X=(np.random.random((N,D))-0.5)*10   #(b - a) * random_sample() + a  :between and a

true_w=np.array([1.0, 0.5,-0.5]+[0]*(D-3))

Y = np.round(sigmoid(X.dot(true_w) + np.random.randn(N)*0.5))
costs=[]
w=np.random.randn(D)/np.sqrt(D)

alpha_=0.001
learning_rate=0.001
l1=3.0   #higher L1 killing more noise, but over 10 does kill signal too
epsilon=1e-10

for t in range(5000):
    Yp = sigmoid(X.dot(w))
    delta=Yp-Y
    w=w-alpha_*((np.dot(X.T,delta))+l1*np.sign(w))
    cost=np.sum((-1.0*(Y*np.log(Yp+epsilon)+(1.0-Y)*np.log(1-Yp+epsilon))))+np.sum(L1*np.abs(w))
    costs.append(cost)
    
plt.plot(costs)
plt.show()

plt.plot(true_w,label='true_w')
plt.plot(w, label='w ')
plt.legend()
```
