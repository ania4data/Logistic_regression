```python
def col_mean_null_replace(x,y):
    
    if pd.isnull(x):
        
        
        if y==1:
            
            return 38.233440860215055            #train[train['Pclass']==1]['Age'].mean()
    
        if y==2:
            
            return 29.87763005780347              #train[train['Pclass']==2]['Age'].mean()    
        
        if y==3:
            
            return 25.14061971830986              #train[train['Pclass']==3]['Age'].mean()
    
    else:
        
        return x
```

```python
train['Age']=train.apply(lambda row: col_mean_null_replace(row['Age'],row['Pclass']),axis=1)
train.head(10)
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='coolwarm')
```

```python
train.drop('Cabin',axis=1,inplace=True)
```

```python
train.head()
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='coolwarm')
```

```python
sex=pd.get_dummies(train['Sex'],drop_first=True)
sex
```

```python
pd.get_dummies(train['Embarked']).head()
```

```python
embark=pd.get_dummies(train['Embarked'],drop_first=True)
```

```python
embark.head()
```

```python
train=pd.concat([train,sex,embark],axis=1)
```

```python
train.head()
```

```python
train.drop(['Sex','Embarked','Name','Ticket','PassengerId'],axis=1,inplace=True)
```

```python
train['Parch'].unique()
```

```python
train.head()
```

```python
train.columns
```

```python
X=train[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'male', 'Q',
       'S']]
Y=train['Survived']
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(X_train,Y_train)
```

```python
predict=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,predict))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(Y_test,predict))
```

```python
predictt=logmodel.predict(X_train)
```

```python
print(classification_report(Y_train,predictt))
```

```python
print(confusion_matrix(Y_train,predictt))
```
